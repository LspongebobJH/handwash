```python
# test rotation
sample = data[0, :, 10, :, :].squeeze(-1)
sample = sample[:, :21]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
sample = (sample - sample.min(axis=1, keepdims=True)) / (sample.max(axis=1, keepdims=True) - sample.min(axis=1, keepdims=True))
ax.scatter(sample[0], sample[1], sample[2])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.view_init(90, 0)
```

```python
from scipy.spatial.transform import Rotation as R
```

```python
r = R.from_euler('z', 0, degrees=True)
sample_1 = r.apply(sample.T).T

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(sample_1[0], sample_1[1], sample_1[2])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.view_init(90, 0)
```
